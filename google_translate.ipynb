```python
import requests
import pandas as pd
import time
from tqdm import tqdm
```

# Load corrected FLORES devtest data

```python
corrected_data = pd.read_csv("Corrected_FLORES_devtest.csv")
corrected_data
```

```python
# convert the corrected data to a dictionary
data = corrected_data[['English', 'Tamazight (Corrected)']].to_dict('list')
# rename the columns
data = {'en': data['English'], 'ber': data['Tamazight (Corrected)']}
```

```python
def translate(sl, tl, s):
	return requests.get(f"https://translate.googleapis.com/translate_a/single?client=gtx&dt=t&sl={sl}&tl={tl}&q={s}").json()[0][0][0]

def get_translations(data, sl, tl, save_to_txt=True):
    translations = data.copy()
    translations['Translation'] = []
    source_texts = data[sl]
    filename = f"gt_{sl}_to_{tl}.txt"

    for source_text in tqdm(source_texts):
        try:
            translation = translate(sl, tl, source_text)
        except Exception as e:
            print(f"Error: {e}")
            translation = None

        translations['Translation'].append(translation)

        if save_to_txt:
            with open(filename, 'a') as f:
                if translation:
                    f.write(translation + '\n')
                else:
                    f.write("\n")

        time.sleep(1)

    translations = pd.DataFrame(translations)
    # replace "sl" with "Source" and "tl" with "Target" in the column names
    translations = translations.rename(columns={sl: 'Source', tl: 'Target'})
    # reorder the columns to have "Source", "Target", and "Translation" as the first three columns
    translations = translations[['Source', 'Target', 'Translation']]

    return translations
```

```python
en_to_ber = get_translations(data, 'en', 'ber')
en_to_ber
```
